```python
from itertools import product


def build_h(m: int):
    n = (1 << m) - 1
    cols = [[(x >> i) & 1 for i in range(m)] for x in range(1, n + 1)]
    return [list(row) for row in zip(*cols)]


def xor_vec(a, b):
    return [x ^ y for x, y in zip(a, b)]


def vector_to_str(v):
    return ''.join(map(str, v))


def wt(v):
    return sum(v)


def print_dual_hamming_simplex(m: int):
    H = build_h(m)
    n = len(H[0])
    expected_weight = 1 << (m - 1)

    print(f"\n=== Симплекс-код (дуальный к коду Хэмминга) для m = {m} ===")
    print(f"  Длина кода:          n = {n}")
    print(f"  Размерность:         k = {m}")
    print(f"  Ожидаемый ненулевый вес слов:      {expected_weight}")
    print(f"  Количество ненулевых слов: { (1 << m) - 1 }\n")

    print("Проверочная матрица H:")
    for i, row in enumerate(H, 1):
        print(f"  Строка {i}:  {vector_to_str(row)}")
    print()

    codewords = []
    for coeffs in product([0, 1], repeat=m):
        if all(c == 0 for c in coeffs):
            continue

        code = [0] * n
        for i, coef in enumerate(coeffs):
            if coef:
                code = xor_vec(code, H[i])

        w = wt(code)
        codewords.append((coeffs, code, w))

    codewords.sort(key=lambda x: (x[2], x[0]))

    print("Все ненулевые кодовые слова симплекс-кода:")
    print("-" * 65)
    for idx, (coeffs, code, w) in enumerate(codewords, 1):
        coeffs_str = vector_to_str(coeffs)
        code_str   = vector_to_str(code)
        print(f"{idx:2d})  coeff: {coeffs_str}   →   слово: {code_str}   вес = {w}")
    print("-" * 65)

    unique_weights = {w for _, _, w in codewords}
    print(f"\nВсего кодовых слов: {len(codewords) + 1}")
    print(f"Уникальные ненулевые веса: {sorted(unique_weights)}")

    if len(unique_weights) == 1 and expected_weight in unique_weights:
        print("\nВЫВОД:Это действительно бинарный симплекс-код!")
    else:
        print("\n ВЫВОД: Свойство симплекс-кода НЕ выполнено.")


if __name__ == "__main__":
    print_dual_hamming_simplex(3)
    # print_dual_hamming_simplex(4)   # раскомментируй, если хочешь увидеть 15 слов
```
